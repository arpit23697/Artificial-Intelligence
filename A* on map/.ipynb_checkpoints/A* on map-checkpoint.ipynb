```python
import pandas as pd
import numpy as np
import math
from collections import defaultdict
import queue
import heapq
```

```python
#this block is to declare the necessary data structures and then take the whole graph as the input

#taking the number of nodes and then the coordinate as the input
n = -1                        #this is for the number of nodes
nodes = dict()                    #storing each node and there coordinate


file = open ("in.txt" , "r")

#this is for the number of nodes n
words = file.readline()
words = words.strip().split()
n = int (words[0])

#this is for reading the nodes and then there coordinates
#each node and there and there coordinate are stored in the dictionary
for _ in range(n):
    line = file.readline()
    words = line.strip().split()
    temp , coordinated = int(words[0]) , (int(words[1]) , int (words[2]))
    temp -= 1
    nodes[temp] = coordinated
    
#taking the number of edges
words = file.readline().strip().split()
edges = int(words[0])

#storing the edges in the adjacency list
graph = [[] for _ in range(n)]
for _ in range(edges):
    
    #taking each edge as the input
    line = file.readline()
    words = line.strip().split()
    x , y, weight = int (words[0]) - 1 , int (words[1]) - 1 , float(words[2])

    #storing each edge in the adjacency list
    graph[x].append( (y , weight))
    graph[y].append( (x , weight))
    
#taking the start state
start = file.readline().strip().split()
start = int(start[0]) - 1

goal = file.readline().strip().split()
goal = int(goal[0]) - 1
    
file.close()
    
    
```

```python
#this block is to print all the informations and then 
print("number of nodes : " , n)
print("Nodes with there coordinates : ")
print(nodes)

print("Graph : ")
for i in range(n):
    print(i+1, end = " ")
    print(graph[i])


print("Start state : " , start)
print("Goal state : " , goal)
```

```python
#this is use the heuristic
def euclidean (x):
    return math.sqrt( (nodes[goal][0] - x[0]) **2 + (nodes[goal][1] - x[1]) ** 2)
```

```python
#solving the problem using the A* approach
def solve():
    d = dict()                    #this is to store the states of each node
                              #first element is the parent
                              #second element is the cost till now
                              #third element is the heuristic
    
    explored = set()          #this is to mark the set of the visible states
    
    counter_queue = 0         #this is to put elements in the queue to break the tie
    
    q = []                    #this is the queue
    heapq.heapify(q)
    
    #initialise all the data structures to be used
    
    for i in range(1 , n):
        d[i] = (-1 , math.inf , euclidean(nodes[i]))
    
    d[start] = (-1 , 0 , euclidean(nodes[start]))
    heapq.heappush (q , ( d[start][1] + d[start][2] , counter_queue , start)) #first element in the queue is estimated cost
                                                                        #second element is to break the tie
                                                                       #third element is the node    
    ans = False
    counter = 0
    while len(q) >= 0 and ans == False:
            counter += 1
            print("Number of steps : " , counter)
            
            #extracting each element
            dist , waste , current = heapq.heappop(q)
            
            #if the extracted node is the goal node then stop the process
            if (current == goal):
                ans = True
                break
            #marking the explored set
            explored.add (current)
            
            #this is for the lazy relaxation
            if (dist > d[current][1] + d[current][2]):
                continue
                
            
            #extracting each edge and then updating the queue
            for v in graph[current]:
                    vertex = v[0]
                    weight = v[1]
                    if (vertex not in explored and d[vertex][1] > d[current][1] + weight   ):
                        
                        d[vertex] = (current , d[current][1] + weight , euclidean(nodes[vertex]) )
                        counter_queue += 1
                        heapq.heappush (q , (d[vertex][1] + d[vertex][2] , counter_queue , vertex))
            
            
    if (ans == False):
        print ("No solution found ")
        return ;

    current = goal
    path = []
    while (current != -1):
        path.insert(0 , current)
        current = d[current][0]
    
    print("The path is : ")
    for i in path:
        print(i+1 , end = " ")
    
    print("\nTotal cost : " , end = " ")
    print(d[goal][1])
    
            
```

```python
solve()
```
