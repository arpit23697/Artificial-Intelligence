```python
#this is to import the function
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import defaultdict
import queue
```

```python

#class for the environment of the n^2 -1 puzzle
#this is also going to have the function for the parity check
class puzzle : 
    def __init__(self):
        n = int(input("Enter the dimension of the matrix : "))
        self.dimension = n
        self.mat = np.zeros((self.dimension , self.dimension))
        self.x = -1
        self.y = -1
        self.goal = np.arange(1,n*n+1).reshape((n , n))
    #this is to take the matrix as the input 
    #in order to take the blank we take * as the input
    def enter_matrix(self):
        print("Enter the elements of the matrix : ")
        for i in range(self.dimension):
            temp  = input().strip().split()
            for j , x in zip(range(self.dimension) , temp):
                if (x == '*'):
                    self.mat[i , j] = self.dimension * self.dimension
                    self.x = i
                    self.y = j
                else:
                    self.mat[i , j] = int(x)
    
    
    def duplicate(self , matrix):
            self.mat = np.copy(matrix)
            for i in range(self.dimension):
                for j in range(self.dimension):
                    if (self.dimension * self.dimension == self.mat[i , j]):
                        self.x = i
                        self.y = j
    #this is the function to define the movements of the blank space
    def left (self):
        if (self.y == 0):
            return False
        else:
            self.mat[self.x , self.y] = self.mat[self.x, self.y - 1]
            self.mat[self.x, self.y - 1] = self.dimension * self.dimension
            self.y = self.y - 1
            return True
        
    def right(self):
        if (self.y == self.dimension - 1):
            return False
        else:
            self.mat[self.x , self.y] = self.mat[self.x , self.y + 1]
            self.mat[self.x , self.y + 1] = self.dimension * self.dimension
            self.y = self.y + 1
            return True
    
    def up(self):
        if (self.x == 0):
            return False
        else:
            self.mat[self.x , self.y] = self.mat[self.x - 1 , self.y]
            self.mat[self.x - 1  , self.y ] = self.dimension * self.dimension
            self.x = self.x - 1
            return True
    
    def down(self):
        if (self.x == self.dimension - 1):
            return False
        else:
            self.mat[self.x , self.y] = self.mat[self.x + 1 , self.y]
            self.mat[self.x + 1, self.y ] = self.dimension * self.dimension
            self.x = self.x  + 1
            return True
    #this is the function to check the parity of the function
    def parity(self):
        s = (self.dimension - self.x - 1) + (self.dimension - self.y - 1)
        for p1 in range(self.dimension * self.dimension - 1):
            #print(s)
            for p2 in range(p1+1 , self.dimension*self.dimension ):
                #extracting the position from p1 and p2
                x1 = p1 // self.dimension
                y1 = p1 % self.dimension
                x2 = p2 // self.dimension
                y2 = p2 % self.dimension

                if (self.mat[x1 , y1 ] > self.mat[x2, y2]):
                    s += 1

                
        print(s)
        #returning the parity value
        if (s % 2 == 0):
            return 0;
        else:
            return 1;
                    
```

```python
#initialising the value and then taking the matrix as the input
#this is to take the matrix and then make it reach the goal state
env = puzzle()
env.enter_matrix()
print("The matrix is : ")
print(env.mat)
```

```python
env.parity()
```

```python
def BFS2():
    if (env.parity() == 1):
        print("No solution")
        return 
    
    d = dict()
    q = queue.Queue()
    q.put(np.copy(env.mat))
    visited = set()
    
    current_string = np.array_str(env.mat)
    visited.add(current_string)
    d[current_string] = ("-1" , "-1")


    ans = False
    if (np.array_equal(env.mat , env.goal)):
        ans = True
        
    counter = 0
    while not q.empty() and ans == False:
        counter+=1
        matrix = q.get()
        matrix_string = np.array_str (matrix)
        env.duplicate(matrix)
        
        print("Number of steps : " , counter)
        #print(matrix)
        #this is to take the new steps
        if (env.left() == True):
            current_string = np.array_str (env.mat)
            if (current_string not in visited):
                q.put(np.copy(env.mat))
                visited.add(current_string)
                d[current_string] = (matrix_string , "left")
                if (np.array_equal(env.mat , env.goal)):
                    ans = True
                    break
            env.right()
        
        if (env.right() == True):
            current_string = np.array_str(env.mat)
            if (current_string not in visited):
                q.put(np.copy(env.mat) )
                visited.add(current_string)
                d[current_string] = (matrix_string , "right")
                if (np.array_equal(env.mat , env.goal)):
                    ans = True
                    break
            env.left()
            
        if (env.up() ==  True):
            current_string = np.array_str (env.mat)
            if (current_string not in visited):
                q.put(np.copy(env.mat))
                visited.add(current_string)
                d[current_string] = (matrix_string , "up")
                if (np.array_equal(env.mat ,env.goal)):
                    ans = True
                    break
            env.down()
        
        if (env.down() ==  True):
            current_string = np.array_str (env.mat)
            if (current_string not in visited):
                q.put(np.copy(env.mat))
                visited.add(current_string)
                d[current_string] = (matrix_string , "down")
                if (np.array_equal(env.mat ,env.goal)):
                    ans = True
                    break
            env.up()
    
  
    if (ans == False):
        print("No Solution")
    else:
        
        current_string = np.array_str(env.mat)
        path = []
        path_matrix = []
        #print("Current string : " , current_string)
        
        while not d[current_string][0] == "-1":
            #path.append(d[current_string][0])
            #path.append(d[current_string][1])
            path.insert(0 , d[current_string][1])
            path_matrix.insert(0 , d[current_string][0])
            current_string = d[current_string][0]
        
        for x , y in zip (path_matrix , path):
            print(x , y)
        print(env.mat)
        
```
