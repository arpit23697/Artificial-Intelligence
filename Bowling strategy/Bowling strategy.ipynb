```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random
```

```python
bowler_stat = [(3,33) , (3.5 , 30 ) , (4 , 24) , (4.5 , 18) , (5,15)]
total_bowlers = 5
overs = 10
wickets = 3
```

```python
#my state space is a dictionary with 3 values
#first parameter is the overs left
#second parameter is the wickets left
#third is how much over each bowler is left with
#that can be encoded as the string

#now i am going to define the mapping for each state
counter = 0
state_to_n = {}
n_to_state = {}


temp = [0 , 1, 2]            #this is for the overs left
                
for over in range(overs + 1):
    for wicket in range(wickets + 1):
        for bowl1 in temp:
            for bowl2 in temp:
                for bowl3 in temp:
                    for bowl4 in temp:
                        for bowl5 in temp:
                            if (bowl1 + bowl2 + bowl3 + bowl4 + bowl5 == over):
                                ans = str(bowl1) + str(bowl2) + str(bowl3) + str(bowl4) + str(bowl5)
                                state = (over , wicket , ans)
                                state_to_n[state] = counter
                                n_to_state[counter] = state 
                                counter+=1
                                
                                
    
```

```python
total_states = counter
print ("Total number of states : " , total_states)
```

```python
def new_bowlstring (string , action):
    temp = int(string[action]) - 1
    ans = ""
    for i in range(action):
        ans = ans + string[i]
        
    
    ans = ans + str(temp)
    
    for i in range(action+1 , 5):
        ans = ans + string[i]
    return ans
```

```python
def value_iteration(value_states):
    policy = np.zeros((total_states , 1))
    threshold = -math.inf
    
    for i in range(total_states):
        state = n_to_state[i]
        bowler_left = state[2]
        over = state[0]
        wicket = state[1]
        
        if (over == 0 or wicket == 0):
            continue
            
        final_values = -math.inf
        temp_value = -math.inf
        for action in [0 , 1, 2,3, 4]:
            if (int(bowler_left[action]) > 0):
                
                temp_value = 0
                
                economy , wicket_take = bowler_stat[action]
                pw = 6.0 / wicket_take
                temp_value = -economy
                
                #if he is going to take the wicket
                bowl_str = new_bowlstring(bowler_left ,action )
                new_state = (over-1 , wicket - 1 , bowl_str)
                
                n1 = state_to_n[new_state]
                temp_value += pw * value_states[n1]
                
                #if he is not going to take wicket
                new_state = (over - 1 , wicket , bowl_str)
                n2 = state_to_n[new_state]
                temp_value += (1 - pw) * value_states[n2]
                
                if (temp_value > final_values):
                    final_values = temp_value
                    policy[i] = action
                    
        
        threshold = max (threshold , abs (value_states[i] - final_values))
        value_states[i] = final_values
        
    return (np.copy(value_states) , np.copy(policy) , threshold)
```

```python
epsilon = 0.000001
total_iteration = 1000
threshold = math.inf
itr = 0

threshold_list = []

value_state = np.zeros ((total_states , 1))

while (threshold > epsilon and itr < total_iteration):
    itr+= 1
    value_state , policy , threshold = value_iteration(value_state)
    threshold_list.append(threshold)
```

```python
def simulate():
    #this is for the simulation of the game
    over_left = 10
    wicket_left = 3
    run = 0
    bowling_str = "22222"

    while (over_left > 0):
        if (wicket_left ==0):
            break



        state = (over_left , wicket_left , bowling_str)
        n = state_to_n[state]

        action = int(policy[n][0])
        economy , wicket_take = bowler_stat[action]
        pw = 6.0 / wicket_take

        bowling_str = new_bowlstring(bowling_str , action)
        run += economy
        r = random.random()
        if (r < pw):
            over_left -= 1
            wicket_left -= 1

        else:
            over_left -= 1
            wicket_left -= 0

        #print ("Over : " , over_left , " wicket : " , wicket_left , "bowler_bowling : " , action , " bowlers : " , bowling_str , " Run : " , run )
    return run
```

```python
total = 0
for itr in range(10000):
    #print ("Game : ", itr)
    total += simulate()
```
