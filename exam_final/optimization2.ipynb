```python
import random 
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
population_size = 10
crossover_prob = 0.2
mutation_prob = 0.1
range_min = -204 
range_max = +204
```

```python
class units:
    def __init__(self , x):
        self.vec = [x[i] for i in range(5)]
        
    def mutation_single (self , x):
        get_bin = lambda x, n: format(x, 'b').zfill(n)
        str1 = get_bin (x , 10)
        ans = ""
        for i in range (len(str1)):
            r = random.random()
            if (r < mutation_prob):
                if (str1[i] == '0'):
                    ans = ans + '1'
                elif (str1[i] == '1'):
                    ans = ans + '0'
                else: 
                    ans = ans + str1[i]
            else:
                ans = ans + str1[i]
        return int(ans , 2)
    
    def mutation (self):
        x_final = []
        for i in range(5):
            a = self.mutation_single(self.vec[i])
            x_final.append(a)     
        return x_final
    
    def crossover_single (self , x1 , x2):
        get_bin = lambda x, n: format(x, 'b').zfill(n)
        str1 = get_bin (x1 , 10)
        str2 = get_bin (x2 , 10)

        r = random.randint(1 ,8)
        child1 = ""
        child2 = ""

        for i in range(r):
            child1 = child1 + str1[i]
            child2 = child2 + str2[i]

        for i in range (r , len (str1)):
            child1 = child1 + str2[i]
            child2 = child2 + str1[i]

        a = int (child1 , 2)
        b = int (child2 , 2)
        return (a , b)
    
    def crossover (self , other):
        x_final1 = []
        x_final2 = []
        for j in range(5):
            a , b = self.crossover_single(self.vec[j] , other.vec[j])
            x_final1.append(a)
            x_final2.append(b)

        return (x_final1 , x_final2)
    
    def fun1 (self):
        y = [self.vec[i] for i in range(5)]
        for i in range(5):
            y[i] = self.vec[i] / 100

        total = 0
        for i in y:
            total += i*i

        return total
    
    def fun2 (self):
        y = [self.vec[i] for i in range(5)]
        for i in range(5):
            y[i] = self.vec[i] / 100

        total = 0
        for i in y:
            total += math.floor(i)

        return total
    
    def fun3 (self):
        y = [self.vec[i] for i in range(5)]
        for i in range(5):
            y[i] = self.vec[i] / 100

        total = 0
        for j , i in zip(range(1 , 6) , y):
            total += j * i**4

        total += random.normalvariate(0 , 1 )

        return total
    
    def __lt__(self , other):
        fx1 = self.fun1()
        fx2 = other.fun1()
        gx1 = self.fun2()
        gx2 = other.fun2()
        hx1 = self.fun3()
        hx2 = other.fun3()

        count1 = 0
        count2 = 0
        if (fx1 < fx2):
            count1+=1
        else:
            count2 += 1

        if (gx1 < gx2):
            count1+=1
        else:
            count2 += 1

        if (hx1 < hx2):    
            count1 +=1
        else:
            count2 += 1

        #print(count1 , count2)
        if (count1 < count2):
            return True
        else:
            return False
        
    def __gt__ (self , other):
        fx1 = self.fun1()
        fx2 = other.fun1()
        gx1 = self.fun2()
        gx2 = other.fun2()
        hx1 = self.fun3()
        hx2 = other.fun3()

        count1 = 0
        count2 = 0
        if (fx1 < fx2):
            count1+=1
        else:
            count2 += 1

        if (gx1 < gx2):
            count1+=1
        else:
            count2 += 1

        if (hx1 < hx2):    
            count1 +=1
        else:
            count2 += 1
        #print(count1 , count2 , "Greater")
        if (count1 < count2):
            return False
        else:
            return True
```

```python
def population_initialise():
    population = []
    for i in range(population_size):
        temp = []
        for j in range(5):
            temp.append (random.randint(range_min , range_max))
        population.append(units(temp))
    return population
```

```python
import copy
def genetic_algorithm (n_episodes):
    
    
    #initialise the population
    
    population = population_initialise()
    generation = 0
    t1 = []
    t2 = []
    t3 = []
    x_final = []
    for itr in range(n_episodes):

        temp = len(population) - population_size
        population = sorted(population)
        if(itr != 0):
            population = population[temp:]
        
        
        generation = generation + 1
        #print ("Generation " , generation , " : " , population[population_size - 1].vec)
        x_final = population[population_size - 1]
        
        t1.append(x_final.fun1())
        t2.append(x_final.fun2())
        t3.append(x_final.fun3())
        n = int(population_size * 0.1)
        
        
        for i in range(n):
            
            #selecting any two at random
            
            r1 = random.choice(population)
            population.remove(r1)
            r2 = random.choice(population)
            population.append(r1)
            
            #crossover and mutation over the child
            c1 , c2 = r1.crossover(r2)
            child1 = units(c1)
            child2 = units(c2)
        
            child1.vec = child1.mutation()
            child2.vec = child2.mutation()
            
            population.append(child1)
            population.append(child2)
            
            
                
    f1 = x_final.fun1()
    f2 = x_final.fun2()
    f3 = x_final.fun2()
    plt.figure(figsize = (10, 6))
    plt.plot(range(len(t1)) , t1 , label = "Function1")
    plt.plot(range(len(t2)) , t2 , label = "Function2")
    plt.plot(range(len(t3)) , t3 , label = "Function3")
    plt.legend()
    plt.title("epochs : " + str(n_episodes) + "\ncrossover probability " + str(crossover_prob) +
              "\nMutation prob " + str(mutation_prob) + "\npopulation size " + str(population_size) + 
              "\nFunction : " + str(f1) + "\nFunction2 : " + str(f2) + "\nFunction3 : " + str(f3) + 
              "\n" + str(x_final.vec[0] / 100) + " " + str(x_final.vec[1] / 100) + " " + str(x_final.vec[2] / 100)
              + " " + str(x_final.vec[3] / 100) + " " + str(x_final.vec[4] / 100))
    
    
    
```

We have used three different type of genetic algorithms to study the convergence of all these functions

* Here the objective is to minimize the three different functions that take 5-dimension vector as the input
* we are not minimizing the sum of these threee functions because it does not make any sense , so we minimizing 
* using the pareto optimization techniques 

* Here is the description of the slightly varying all the genetic algorithms

Basic genetic algorithm
 * We crossover by random selecting the 10% of the population and put their two childs in the population
 * After crossover and mutation ,since we have put the child in the population so the size of the population
     will increase. So we sort the population according to fitness and then take the best
     population_size of the population
  
Diversity
 * We crossover the first 10% of the population with the last 10% of the population and put their two childs in the 
     population
 * After crossover and mutation ,since we have put the child in the population so the size of the population
     will increase. So we sort the population according to fitness and then take the best
     population_size of the population
  

Elitism
  * We take the 10% of the fittest population put it in the elite group
  * Then we do the crossover by in the remaining population and their childs two the remaining population
  * After crossover and mutation ,since we have put the child in the population so the size of the population
     will increase. So we sort the population according to fitness and then take the best
     (population_size - elite_group_size) of the population
  * Then we add the elite group to population.
  * then we sort the population again according to the fitness and run the next iteration
  * In short elite group will be transfered to the next generation
  
Random search
  * In this simply we find the best entry from the randomly generated population
