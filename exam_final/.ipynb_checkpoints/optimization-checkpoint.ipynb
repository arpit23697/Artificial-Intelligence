```python
import random 
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
population_size = 100
crossover_prob = 0.2
mutation_prob = 0.1
range_min = -204 
range_max = +204
```

```python
def mutation_single (x):
    get_bin = lambda x, n: format(x, 'b').zfill(n)
    str1 = get_bin (x , 10)
    ans = ""
    for i in range (len(str1)):
        r = random.random()
        if (r < mutation_prob):
            if (str1[i] == '0'):
                ans = ans + '1'
            elif (str1[i] == '1'):
                ans = ans + '0'
            else: 
                ans = ans + str1[i]
        else:
            ans = ans + str1[i]
    return int(ans , 2)
```

```python
def mutation (x):
    x_final = []
    for i in range(5):
        a = mutation_single(x[i])
        x_final.append(a)     
    return x_final
```

```python
a_ = [0 , 10 , 20 , 30 , 50]
mutation(a_)
```

```python
def crossover_single (x1 , x2):
    get_bin = lambda x, n: format(x, 'b').zfill(n)
    str1 = get_bin (x1 , 10)
    str2 = get_bin (x2 , 10)
    
    r = random.randint(1 ,8)
    child1 = ""
    child2 = ""
    
    for i in range(r):
        child1 = child1 + str1[i]
        child2 = child2 + str2[i]
        
    for i in range (r , len (str1)):
        child1 = child1 + str2[i]
        child2 = child2 + str1[i]
    
    a = int (child1 , 2)
    b = int (child2 , 2)
    return (a , b)
```

```python
def crossover (x1 , x2):
    x_final1 = []
    x_final2 = []
    for j in range(5):
        a , b = crossover_single(x1[j] , x2[j])
        x_final1.append(a)
        x_final2.append(b)
        
    return (x_final1 , x_final2)
```

```python
def fun1 (x):
    y = [x[i] for i in range(5)]
    for i in range(5):
        y[i] = x[i] / 100
        
    total = 0
    for i in y:
        total += i*i
        
    return total
```

```python
def fun2 (x):
    y = [x[i] for i in range(5)]
    for i in range(5):
        y[i] = x[i] / 100
        
    total = 0
    for i in y:
        total += math.floor(i)
        
    return total
```

```python
def fun3 (x):
    y = [x[i] for i in range(5)]
    for i in range(5):
        y[i] = x[i] / 100
        
    total = 0
    for j , i in zip(range(1 , 6) , y):
        total += j * i**4
        
    total += random.normalvariate(0 , 1 )
        
    return total
```

```python
def fitness(x1 , x2):
    fx1 = fun1 (x1)
    fx2 = fun1 (x2)
    gx1 = fun2 (x1)
    gx2 = fun2 (x2)
    hx1 = fun3 (x1)
    hx2 = fun3 (x2)
    
    count1 = 0
    count2 = 0
    if (fx1 < fx2):
        count1+=1
    else:
        count2 += 1
        
    if (gx1 < gx2):
        count1+=1
    else:
        count2 += 1
    
    if (hx1 < hx2):    
        count1 +=1
    else:
        count2 += 1
    
    if (count1 > count2):
        return x1
    else:
        return x2
```

```python
a_ = [1 , 20 , 30 , 200 , 200]
b_ = [1 , 20 , 20 , 10 , 201]
fitness(a_ , b_)
```

```python
def best (L):
    ans = L[0]
    for i in range(len(L)):
        ans = fitness(ans , L[i])
    
    L.remove(ans)
    ans2 = L[0]
    for i in range(len(L)):
        ans2 = fitness(ans2 , L[i])
    L.remove(ans2)
    return (ans , ans2)
    
    
```

```python
def population_initialise():
    population = []
    for i in range(population_size):
        temp = []
        for j in range(5):
            temp.append (random.randint(range_min , range_max))
        population.append(temp)
    return population
```

```python
import copy
def genetic_algorithm (n_episodes):
    
    
    #initialise the population
    
    population = population_initialise()
    generation = 0
    t1 = []
    t2 = []
    t3 = []
    x_final = []
    for itr in range(n_episodes):

        random.shuffle (population)
        #print (population)
        
        generation = generation + 1
        x_final = population[0]
        
        for i in range(population_size):
            x_final = fitness (x_final , population[i])
        
        #print ("Generation " , generation , " : " , x_final)
        
        t1.append(fun1(x_final))
        t2.append(fun2(x_final))
        t3.append(fun3(x_final))
        n = population_size //2
        
        
        for i in range(n):
            r = random.random()
            if (r < crossover_prob):
                x , y = crossover (population[i] , population[i+n])
                x = mutation(x)
                y = mutation(y)
                L = [x , y , population[i] , population[i+n]]
                ans1 , ans2 = best(L)
                
                population[i] = ans1
                population[i+n] = ans2
                
    f1 = fun1 (x_final)
    f2 = fun2 (x_final)
    f3 = fun3 (x_final)
    plt.figure(figsize = (10, 6))
    plt.plot(range(len(t1)) , t1 , label = "Function1")
    plt.plot(range(len(t2)) , t2 , label = "Function2")
    plt.plot(range(len(t3)) , t3 , label = "Function3")
    plt.legend()
    plt.title("epochs : " + str(n_episodes) + "\ncrossover probability " + str(crossover_prob) +
              "\nMutation prob " + str(mutation_prob) + "\npopulation size " + str(population_size) + 
              "\nFunction : " + str(f1) + "\nFunction2 : " + str(f2) + "\nFunction3 : " + str(f3) + 
              "\n" + str(x_final[0] / 100) + " " + str(x_final[1] / 100) + " " + str(x_final[2] / 100)
              + " " + str(x_final[3] / 100) + " " + str(x_final[4] / 100))
    
    
    
```

```python
population = [50, 100 , 200]
cp = [0.2 , 0.3]
mp = [0.01 , 0.05]
n_episode = [500]
```

```python
for x in population:
    for y in cp:
        for z in mp:
            for a in n_episode:
                population_size = x
                crossover_prob = y
                mutation_prob = z
                genetic_algorithm(a)
```

```python
def genetic_random():
    population = []
    f1 = []
    f2 = []
    f3 = []
    for i in range(population_size):
        temp = []
        for j in range(5):
            temp.append (random.randint(range_min , range_max))
        population.append(temp)
        f1.append(fun1(temp))
        f2.append(fun2(temp))
        f3.append(fun3(temp))
    
    x_final = population[0]
    for i in range(population_size):
        x_final = fitness(x_final , population[i])
        
    plt.figure(figsize = (10, 6))
    t1 = fun1(x_final)
    t2 = fun2(x_final)
    t3 = fun3(x_final)
    plt.plot(range(len(f1)) , f1 , label = "Function1")
    plt.plot(range(len(f2)) , f2 , label = "Function2")
    plt.plot(range(len(f3)) , f3 , label = "Function3")
    plt.title("population size " + str(population_size) + 
              "\nFunction : " + str(t1) + "\nFunction2 : " + str(t2) + "\nFunction3 : " + str(t3) + 
              "\n" + str(x_final[0] / 100) + " " + str(x_final[1] / 100) + " " + str(x_final[2] / 100)
              + " " + str(x_final[3] / 100) + " " + str(x_final[4] / 100))
```

We have used three different type of genetic algorithms to study the convergence of all these functions

* Here the objective is to minimize the three different functions that take 5-dimension vector as the input
* we are not minimizing the sum of these threee functions because it does not make any sense , so we minimizing 
* using the pareto optimization techniques 

* Here is the description of the slightly varying all the genetic algorithms

Basic genetic algorithm
 * We crossover the first half of the population with the second half of the population
 * After crossover we have two more childs
 * so Among the two parents and the two childs we select the best two and replace them with the parents
  i.e. keep the best two in the population
  
Diversity
 * We crossover the first half of the population with the second half of the population
 * After crossover we have two more childs
 * so Among the two parents and the two childs we select the two randomly and replace them with the parents

Elitism
  * we crossover between the two fittest parents
  * After crossover we have two more childs
  * so Among the two parents and the two childs we select the best two and replace them with the parents

Random search
  * In this simply we find the best entry from the randomly generated population
